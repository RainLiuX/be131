# Lab 6

In this lab section, I am provided with the MiSeq reads of *Shewanella oneidensis*. However, the data may be **contaminated with human genome**. In order to avoid that, we will first **align the reads with human genome**, and filter out any aligned sequences. The we **map the "washed" data to the genome of *S.oneidensis*** to see the depth of each region. The MiSeq results is likely to be the RNA sequencing of some transcriptomic assay. The depths of each region reveals of the expression level of that region, which is very helpful if the genome is somewhat annotated. Or it is just genome sequencing.

## Align the reads to the human genome

First, locate the MiSeq data/genome data on the server. The path for sequence reads is at `/data/lab6/illumina_reads.fastq`, and the database for human genome is at `/data/lab6/shewanella_oneidensis_mr-1`

Bowtie is the commandline tool that maps short sequences to th genome. Use the following command to filter the sequences that cannot be aligned with human genome:

```sh
bowtie2 -p 4 --very-fast --no-unal -x /data/hg19/hg19 -U /data/lab6/illumina_reads.fastq -S human_align.sam --un human_unaligned.fastq 2> human_align.stdout
```
The `--no-unal` parameter tells bowtie not to put unaligned sequences into SAM file, in order to conpress the file size. The `--un` parameter designates the file to dump all unaligned sequences, which are the "washed" sequences we want.

For some reason, the output of bowtie2 is in `stderr` instead of `stdout`. One have to use `2>` instead of `1>` to save the output.

Let's have a look at how much contamination is here:

````python
human_align_output = open('human_align.stdout')
for line in human_align_output :
    print(line)
````

**8.57%** of the sequence aligns to human genome.

## Align the reads to the *S.oneidensis* genome

Then use the unaligned sequences in the last step to map the *S.oneidensis* genome. Such as:

```sh
bowtie2 -p 4 --very-fast --no-unal -x /data/lab6/shewanella_oneidensis_mr-1 -U human_unaligned.fastq -S s.o_align.sam --un s.o_unaligned.fastq 2> s.o_align.stdout
```

Now we have the alignment data of the MiSeq to the bacteria genome, without the contamination of human genome sequence. The next step would be examining the mapping result. Before, we shall check the alignment rate:

````python
human_align_output = open('s.o_align.stdout')
for line in human_align_output :
    print(line)
````

**Over 70%** of the sequences match the bacteria genome. Would the matching rate be further increased if we use `--very-sensitive` parameter to get a slower but more accurate outcome?

```sh
bowtie2 -p 4 --very-sensitive --no-unal -x /data/lab6/shewanella_oneidensis_mr-1 -U human_unaligned.fastq -S s.o_align_sensitive.sam --un s.o_unaligned_sensitive.fastq 2> s.o_align_sensitive.stdout
```

````python
human_align_output = open('s.o_align_sensitive.stdout')
for line in human_align_output :
    print(line)
````

As we can see, using the `--very-sensitive` parameter, the alignment rate **increased by 0.14%**. This is not a significant number. Or we can say, that it is not necessary to use "very sensitive" mode in stead of speed mode.

However, there is still about 30% of the sequences that cannot be aligned to the bacteria genome. This could due to the contamination of other organisms, or these sequences could undergo some rare post-transcriptional modifications.

## Generate a coverage plot

Which part of the genome has more or less match then other? To answer this question, we need to get the **"depth"** of each position from the SAM (sequence alignment map) file we have just generated. **Samtools** is a useful commandline tool in processing sam files. First, we can convert the sam file to a binary format bam file:

```sh
samtools view -b s.o_align.sam s.o_align.bam
```

After the compression, the file size dropped from **440MB** to **176MB**. Then we sort and index the bam file:

```sh
samtools sort s.o_align.bam > s.o_align.sorted.bam
samtools index s.o_align.sorted.bam
```

Thses steps would give a correct coordinate to the alignment map. At last, generate the depth of the map:

```sh
samtools depth -a s.o_align.sorted.bam > s.o_pileup.tab
```

The `-a` parameter will make the output file include those positions that has no coverage at all. Finally, use python to read and proceed this file:

````python
%matplotlib inline
import sys, numpy
import matplotlib.pyplot as plt
in_file = open('s.o_pileup.tab')
depths = []
for line in in_file.readlines() :
    depths.append(int(line.split('\t')[-1][:-1]))
print('Max depth:', max(depths))
print('Min depth:', min(depths))
print('Average depth', numpy.mean(depths))
````

````python
fig = plt.figure(figsize=(20,10), dpi=100)
axes = fig.add_subplot(111)
_ = axes.plot(range(1, len(depths)+1), depths, color='red')
````

````python
fig = plt.figure(figsize=(20,10), dpi=100)
axes = fig.add_subplot(111)
_ = axes.hist(depths, 100)
````

Here, we can further information about the MiSeq map. We can see that the max depth is 269 and the average depth is about 37. We can roughly get the idea from which region in the genome has a higher or lower coverage.
The histogram is also shown here, to demostrate a rather good **normal distribution** of the coverage number of each position.

## Extra credit 1: What is the sex of jamie?

Almost all humans have a pair of sex chromosomes, which can be used to judge one's sex. We assume that all human genome contamination are from Jamie. According to the match number to the X and/or Y chromosome, we can speculate Jamie's sex.

Fist, we need to process the sequence match to the human genome just like we did for the S.oneidensis* genome map data:

```sh
samtools view -b human_align.sam > human_align.bam
samtools sort human_align.bam > human_align.sort.bam
samtools index human_align.sorted.bam
samtools depth human_align.sort.bam > human_pileup.tab
```

Remember not to use -a parameter in the last step, in order not to get a super big file.

````python
in_file = open('human_pileup.tab')
human_depths = []
for line in in_file.readlines() :
    human_depths.append(int(line.split('\t')[-1][:-1]))
````

````python
fig = plt.figure(figsize=(20,10), dpi=100)
axes = fig.add_subplot(111)
_ = axes.plot(range(1, len(human_depths)+1), human_depths, color='red')
````

````python
in_file = open('human_pileup.tab')
# Get the tags of chromosomes
chro_list = ['chr1', 'chr2', 'chr3', 'chr4', 'chr5', 'chr6', 'chr7',\
            'chr8', 'chr9', 'chr10', 'chr11', 'chr12', 'chr13', 'chr14',\
            'chr15', 'chr16', 'chr17', 'chr18', 'chr19', 'chr20', 'chr21',\
            'chr22', 'chrX', 'chrY']
chro_depth = {}
for key in chro_list :
    chro_depth[key] = 0
for line in in_file :
    line_list = line.split('\t')
    try:
        chro_depth[line_list[0]] += int(line_list[2])
    except KeyError:
        break
````

## Extra Credit 2: Zoom in the High Coverage Area

Genes of high coverage (sometimes low coverage) are the regions that we have great insterest in, since if we consider this is the cDNA sequencing file which reflects the mRNA level, then high coverage regions suggest a higher expression of that particular gene, which make us interest.

To find the "high" coverage region, we can first define a threshold, values above this value is counted as "high". Say the __threshold__ is __400%__ as large as the average.

````python
fig = plt.figure(figsize=(20,10), dpi=100)
axes = fig.add_subplot(111)
depth_norm = []; depth_thre = []
depth_avg = sum(depths) / len(depths)
threshold = 4
for depth in depths :
    if depth > depth_avg * threshold :
        depth_thre.append(depth)
    else:
        depth_norm.append(depth)
print(len(depth_thre))
````
