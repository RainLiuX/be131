```python
%matplotlib inline
import sys, re, math
import matplotlib.pyplot as plt
infile = open('romeoandjuliet.txt', 'r')
wordsDict = {}
for line in infile:
    line = line.lower().rstrip()
    words = re.findall(r"[\w']+", line)
    for word in words:
        if word in wordsDict:
            wordsDict[word] += 1
        else:
            wordsDict[word] = 1
#for word in sorted(wordsDict.keys()):
#    print(word, ":", wordsDict[word])
fig = plt.figure(figsize=(40,20), dpi=100)
sortedKey = sorted(wordsDict, key=wordsDict.get, reverse=True)
print("10 most frequent words:")
for i in range(10):
    print(sortedKey[i], wordsDict[sortedKey[i]])
print("")
sortedDict = sorted(wordsDict.values(), reverse=True)
print("All words exist only once:")
fre1words=[]
for word in wordsDict:
    if wordsDict[word] == 1:
        fre1words.append(word)
print(*fre1words)

top = 50
plt.bar(range(1,top), sortedDict[:top-1], axes=fig.add_subplot(121))
# plotting in log scale
logX = range(math.floor(math.log10(len(sortedDict)))+1)
plt.bar(logX, [math.log10(sortedDict[10**i]) for i in logX], axes=fig.add_subplot(122))
```
